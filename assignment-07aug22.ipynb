```r
 #Que5) x <- c('Bombay', 'Mumbai', 'delhi', 'gandhinagar', 'Chennai')
## Find all the cities which starts with capital letters.
##Find all the cities which starts with small letters.

#Answer-
 x <- c('Bombay', 'Mumbai', 'delhi', 'gandhinagar', 'Chennai')
x
grep("^[A-Z]",x) #it returns index of matching first character
grep("^[A-Z]",x,value=T) #it returns values of matching first character
grep("^[a-z]",x) #same like above(index)
grep("^[a-z]",x,value=T) #same like above(values)
```
